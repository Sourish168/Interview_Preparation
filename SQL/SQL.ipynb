## Connecting SQL in Jupyter using SQLite3

```python
import sqlite3
import pandas as pd

# Create a database
conn = sqlite3.connect("ecommerce.db")
```

```python
# Load CSV files
path = "./../Datasets/"
customers = pd.read_csv(path + "customers.csv")
products = pd.read_csv(path + "products.csv")
employees = pd.read_csv(path + "employees.csv")
transactions = pd.read_csv(path + "transactions.csv")
students = pd.read_csv(path + "students.csv")

# Write DataFrames into SQL tables
customers.to_sql("customers", conn, if_exists="replace", index=False)
products.to_sql("products", conn, if_exists="replace", index=False)
employees.to_sql("employees", conn, if_exists="replace", index=False)
transactions.to_sql("transactions", conn, if_exists="replace", index=False)
students.to_sql("students", conn, if_exists="replace", index=False)
```

```python
# Execute SQL queries (Example)
query = """
SELECT City, COUNT(*) AS total_customers
FROM customers
GROUP BY City
ORDER BY total_customers DESC;
"""
pd.read_sql(query, conn)
```

## Section 1: SQL Basics (1–20)

#### **1. What is SQL?**
**SQL (Structured Query Language)** is the standard programming language used to manage, manipulate, and query data stored in **Relational Database Management Systems (RDBMS)** (like PostgreSQL, MySQL, SQL Server, and Oracle).

* **Purpose:** It interacts specifically with *structured data* organized into tables (rows and columns).
* **Core Operations (CRUD):** It allows you to **C**reate, **R**ead, **U**pdate, and **D**elete data.
* **Declarative Nature:** You write SQL to tell the database *what* data you want (e.g., `SELECT name FROM employees`), and the database engine figures out *how* to retrieve it efficiently.
* **Key Sub-languages:**
* **DDL (Data Definition Language):** Defines structure (`CREATE`, `ALTER`, `DROP`).
* **DML (Data Manipulation Language):** Modifies data (`INSERT`, `UPDATE`, `DELETE`).
* **DQL (Data Query Language):** Retrieves data (`SELECT`).
* **DCL (Data Control Language):** Manages permissions (`GRANT`, `REVOKE`).
* **TCL (Transaction Control Language):** Manages transactions (`COMMIT`, `ROLLBACK`).

#### **2. What are the different types of SQL commands?**
SQL commands are categorized into five main groups based on their specific functions. Here is a quick reference table perfect for your interview prep notebook:

| Category | Purpose | Key Commands |
| --- | --- | --- |
| **DDL** (Data Definition Language) | Defines or modifies the database structure (schema, tables). | `CREATE`, `ALTER`, `DROP`, `TRUNCATE` |
| **DML** (Data Manipulation Language) | Modifies the actual data (rows) stored within the tables. | `INSERT`, `UPDATE`, `DELETE` |
| **DQL** (Data Query Language) | Retrieves data from the database. | `SELECT` |
| **DCL** (Data Control Language) | Manages user access, rights, and database security. | `GRANT`, `REVOKE` |
| **TCL** (Transaction Control Language) | Manages database transactions to ensure data integrity. | `COMMIT`, `ROLLBACK`, `SAVEPOINT` |

<br>

> **Interview Tip:** While `SELECT` is universally considered **DQL**, some older texts or interviewers might loosely group it under **DML**. If asked, clarify that `SELECT` is strictly DQL because it *queries* data without *modifying* it.

#### **3. What is the difference between DDL, DML, DCL, TCL, and DQL?**
To ace this question in an interview, the best way to explain the difference is by contrasting their **scope of impact** (what they actually change in the database) and their **transactional behavior**.

Here is a simple analogy to remember the difference:

* **DDL** builds the house (structure).
* **DML** moves the furniture in and out (data).
* **DQL** looks through the window to see what's inside (viewing).
* **DCL** decides who gets a key to the front door (security).
* **TCL** acts as the moving insurance, ensuring the furniture move is either fully completed or completely reversed if something goes wrong (state control).

**Key Differences at a Glance**

| Feature | DDL (Definition) | DML (Manipulation) | DQL (Query) | DCL (Control) | TCL (Transaction) |
| --- | --- | --- | --- | --- | --- |
| **Target** | Database Schema / Metadata | Table Rows / Actual Data | Table Data | User Roles / Permissions | DML Operations |
| **Action** | Creates, alters, or destroys structures. | Adds, modifies, or removes records. | Retrieves records. | Grants or revokes access rights. | Commits or rolls back changes. |
| **Auto-Commit?** | **Yes** (Changes are saved permanently immediately).* | **No** (Changes must be explicitly committed using TCL). | N/A (Read-only) | **Yes** | N/A (These *are* the commit commands) |
| **Reversible?** | Cannot be easily rolled back. | Can be rolled back (before a commit). | N/A | Cannot be rolled back. | N/A |

<br>

> **Interview Tip:** Interviewers often test your knowledge of the "Auto-Commit" row. Knowing that a `DROP TABLE` (DDL) is immediately permanent and cannot be rolled back, while a `DELETE FROM` (DML) can be rolled back (if uncommitted), proves you understand how the engine handles these commands differently.
> *(Note: Some RDBMS like SQL Server allow DDL to be rolled back within explicit transactions, but in standard SQL and Oracle, DDL auto-commits).*

#### **4. What is a database?**
A **database** is an organized, structured collection of data stored electronically in a computer system. It is designed to make data management—storing, retrieving, updating, and deleting data—highly efficient and secure.

**Make sure to highlight *why* we use databases over simple files (like Excel):**
* **Scale & Performance:** Built to handle massive amounts of data and complex queries rapidly.
* **Concurrency:** Allows multiple users or applications to access and modify the data at the same time without conflicts or data corruption.
* **Data Integrity:** Enforces rules (like constraints and data types) to ensure the information is accurate and reliable.
* **Security:** Provides strict access controls to determine who can view or change specific data.

> **Interview Tip:** Be careful to distinguish between the **database** (the actual data stored) and the **DBMS (Database Management System)**. The DBMS is the software (like MySQL or Oracle) used to interact with the database.

#### **5. What is a table?**
A **table** is the fundamental structural unit of a relational database. It is a database object used to store data in a highly organized, grid-like format consisting of rows and columns. Think of it as the database equivalent of a single Excel spreadsheet tab, but with strictly enforced rules about what type of data can go where.

**Core Components**
* **Columns (Fields / Attributes):** These represent the vertical structure. Each column defines a specific property of the data (e.g., `Employee_ID`, `Name`, `Salary`) and dictates the **data type** (e.g., integer, text, decimal) that can be stored in it.
* **Rows (Records / Tuples):** These represent the horizontal structure. Each row represents a single, complete, and distinct entry or instance of data (e.g., all the information for one specific employee).
* **Cells:** The intersection of a single row and a single column, holding one specific data value.

> **Interview Tip:** In more rigorous data or business analytics interviews, you might hear formal relational algebra terminology. Remember that mathematically, a table is called a **Relation**, a row is a **Tuple**, and a column is an **Attribute**.

#### **6. What is a row and a column?**
In a relational database, rows and columns form the grid structure of a table. Here is how to distinguish them for an interview:

**Column (Field / Attribute)**
* **Definition:** The vertical component of a table.
* **Function:** It represents a specific property or characteristic of the data (e.g., `First_Name`, `Hire_Date`, `Salary`).
* **Constraint:** Every column is strictly bound by a **data type** (e.g., `INT`, `VARCHAR`, `DATE`). This means every single value entered into that column must match that format.

**Row (Record / Tuple)**
* **Definition:** The horizontal component of a table.
* **Function:** It represents a single, complete, and distinct instance of data.
* **Context:** If you have an `Employees` table, one row holds all the related column information (name, date, salary) for *one specific employee*.

> **Interview Tip:** You can elevate your answer by using the right terminology based on the context. In practical, day-to-day SQL, use **Fields** (columns) and **Records** (rows). If the interview shifts toward database theory or relational algebra (common in business analytics or data science interviews), use **Attributes** (columns) and **Tuples** (rows).

#### **7. What is a primary key?**
A **Primary Key** is a column (or a combination of columns) that uniquely identifies every distinct row in a relational database table. Think of it as the unique ID number for a specific record, like a fingerprint, a passport number, or an employee ID.

**Core Rules of a Primary Key**
* **Uniqueness:** No two rows can have the same primary key value.
* **Not Null:** A primary key cannot contain a `NULL` (missing) value. Every single record *must* have a valid primary key.
* **One per Table:** A table can only have one primary key constraint. However, that single key can be made up of multiple columns working together, which is called a **Composite Key**.
* **Immutable (Best Practice):** While technically possible to change, primary key values ideally should not change over time.

> **Interview Tip:** A very common follow-up question is, *"What is the difference between a Primary Key and a Unique Constraint (Unique Key)?"*
> The key distinction is that a table can have *multiple* Unique Keys, and a Unique Key typically allows exactly *one* `NULL` value. A table can only have *one* Primary Key, and it can *never* contain a `NULL` value.

#### **8. What is a foreign key?**
A **Foreign Key** is a column (or group of columns) in one table that links directly to the **Primary Key** of another table. It acts as a bridge to establish and enforce a relationship between the two tables.

**Core Characteristics**
* **Referential Integrity:** This is the main purpose of a foreign key. It ensures that you cannot enter a value in the child table if that value does not already exist in the parent table. (For example, you cannot assign an employee to `Department_ID = 5` if Department 5 does not exist in the Departments table).
* **Parent-Child Relationship:** The table with the Primary Key is the **Parent** (or referenced) table. The table with the Foreign Key is the **Child** (or referencing) table.
* **Duplicates Allowed:** Unlike primary keys, a foreign key column can contain duplicate values. This is how "one-to-many" relationships work (e.g., multiple employees can belong to the exact same department).
* **NULL Values Allowed:** A foreign key *can* accept `NULL` values. This simply implies that the relationship is optional or unknown at the time (e.g., a new employee who hasn't been assigned to a department yet).

> **Interview Tip:** Interviewers love asking, *"Can a table have more than one Foreign Key?"* and *"Can a Foreign Key be NULL?"* The answer to both is **Yes**. A table can have multiple foreign keys linking to various other tables, and it absolutely can contain `NULL` values (unlike a Primary Key).

#### **9. What is a unique key?**
A **Unique Key** is a database constraint that ensures all data values in a column (or a combination of columns) are entirely distinct from one another. It prevents any duplicate records from being entered into that specific column.

**Core Characteristics**
* **Uniqueness:** No two rows can have the same value in a Unique Key column.
* **NULL Values Allowed:** Unlike a Primary Key, a Unique Key *can* accept `NULL` (missing) values. (Note: How many `NULL`s are allowed can depend on the specific RDBMS—for instance, MySQL allows multiple `NULL`s because `NULL` is not considered equal to another `NULL`, while SQL Server traditionally allows only one).
* **Multiple per Table:** A single table can have multiple Unique Keys. For example, in a `Users` table, you might have a Primary Key on `User_ID`, but put Unique Keys on both the `Email` and `Phone_Number` columns.
* **Index Creation:** When you define a Unique Key, the database engine automatically creates a unique index on that column to quickly enforce the rule during data entry.

> **Interview Tip:** In database design, columns that have a Unique Key constraint are often called **Candidate Keys**. They are columns that *could* potentially serve as the Primary Key (because they uniquely identify a row, like an Email address or Social Security Number), but were not chosen as the main identifier for the table.

#### **10. What is a composite key?**
A **Composite Key** is a key that consists of two or more columns that, when combined, uniquely identify a single row in a table. It is used when a single column is not enough to guarantee the uniqueness of a record on its own.

**Core Characteristics**
* **Combined Uniqueness:** The individual columns making up the key can contain duplicate values, but the *combination* of those values across the row must be entirely unique.
* **Role Versatility:** While most commonly used to create a **Composite Primary Key**, you can also have composite Unique Keys and composite Foreign Keys.
* **Junction Tables:** They are heavily used in "many-to-many" relationships to link two tables together.

**A Practical Example**
Imagine an `Order_Details` table.

* An `Order_ID` alone cannot be the primary key because one order might contain multiple items (duplicate `Order_ID`s).
* An `Item_ID` alone cannot be the primary key because the same item can be sold in many different orders (duplicate `Item_ID`s).
* However, the combination of `(Order_ID, Item_ID)` will be unique for every single row. That combination is your Composite Key.

> **Interview Tip:** If an interviewer asks when you *shouldn't* use a composite key, mention scalability and simplicity. If a composite key requires combining 4 or 5 columns to guarantee uniqueness, it is usually better practice to create a brand new, single-column **Surrogate Key** (like an auto-incrementing ID) to serve as the Primary Key instead.

#### **11. What is a candidate key?**
A **Candidate Key** is a column (or a combination of columns) that meets all the strict requirements to serve as the Primary Key for a table. Think of it as the name implies: it is a "candidate" running for the office of Primary Key.

**Core Characteristics**
* **Uniqueness:** It must uniquely identify every single row in the table.
* **Not Null:** It cannot contain `NULL` (missing) values, because a missing value cannot uniquely identify a record.
* **Minimality:** It must use the minimum number of columns necessary to guarantee uniqueness. If you have a multi-column key, and you can remove one column but the remaining columns still guarantee uniqueness, the original combination is *not* a candidate key (it is known as a **Super Key**).
* **Multiple Candidates:** A single table can have several candidate keys. For example, in an `Employees` table, both the `Employee_ID` and the `Social_Security_Number` could independently serve as candidate keys.

**The Primary vs. Alternate Key Relationship**  
During database design, you look at all your Candidate Keys and choose exactly *one* to be the official **Primary Key**. Any Candidate Keys that were *not* selected are then referred to as **Alternate Keys** (and are usually enforced using a Unique constraint).

> **Interview Tip:** If an interviewer asks you to define the difference between a Super Key and a Candidate Key, use the word **"minimality."** A Super Key is *any* combination of columns that uniquely identifies a row. A Candidate Key is a *minimal* Super Key—meaning it contains no unnecessary extra columns.

#### **12. What is a surrogate key?**
A **Surrogate Key** is an artificially generated, system-assigned identifier used as the Primary Key for a table. Unlike other keys, a surrogate key has absolutely **no business meaning** or connection to the real-world data in the row. It exists purely for the database's internal use to uniquely identify a record.

**Core Characteristics**
* **System-Generated:** It is usually created automatically by the database engine when a new row is inserted (e.g., using `AUTO_INCREMENT` in MySQL, `IDENTITY` in SQL Server, or `SERIAL` in PostgreSQL).
* **Meaningless:** It contains no factual information about the record. For instance, an auto-incrementing integer like `1001`, `1002`, or a UUID string.
* **Immutable:** Because it has no real-world meaning, it never needs to be updated. (If a user changes their email address, a surrogate key remains exactly the same).
* **Performance:** They are almost always simple numeric data types (like `INT` or `BIGINT`), making them highly efficient for indexing and joining tables.

> **Interview Tip:** You will almost always be asked to contrast a Surrogate Key with a **Natural Key**.
> * A **Natural Key** is a column that naturally exists in the real world and guarantees uniqueness (like an Email Address or a Social Security Number).
> * **Why use a Surrogate instead of a Natural?** If you use an Email Address as a Primary Key, and the user changes their email, you have to update that value in the parent table *and* cascade that update to every single child table using it as a Foreign Key. A Surrogate Key prevents this massive update chain because it never changes.

#### **14. What is the difference between CHAR and VARCHAR?**
The fundamental difference between **CHAR** and **VARCHAR** lies in how they allocate **storage space** for strings in the database.
* **`CHAR` (Character):** A fixed-length data type. If you define a column as `CHAR(10)` and insert the word "SQL" (3 letters), the database will pad it with 7 empty spaces to make it exactly 10 characters long.
* **`VARCHAR` (Variable Character):** A variable-length data type. If you define a column as `VARCHAR(10)` and insert "SQL", it will only use 3 characters' worth of storage space (plus 1 or 2 bytes overhead to record the string's length).

**Key Differences at a Glance**

| Feature | CHAR | VARCHAR |
| --- | --- | --- |
| **Length** | Fixed | Variable |
| **Storage** | Uses the exact defined size, padding empty space with blanks. | Uses only the required size + length prefix. |
| **Performance** | Slightly faster (fixed memory blocks are easier to process). | Slightly slower (requires dynamic memory calculation). |
| **Best Used For** | Data with a consistent, known length. | Data where lengths vary significantly. |
| **Examples** | State codes ('WB', 'MH'), Gender ('M', 'F', 'O'), PIN codes. | Names, Emails, Home Addresses, Passwords. |

<br>

> **Interview Tip:** A great nuance to mention is the **comparison trap**. Because `CHAR` pads with spaces, comparing strings can sometimes be tricky depending on the database engine. For example, comparing the `CHAR` value `'Data      '` with the `VARCHAR` value `'Data'` might fail if the engine doesn't automatically trim trailing spaces during the comparison.

#### **15. What is the difference between INT, FLOAT, and DECIMAL?**
The difference between **INT**, **FLOAT**, and **DECIMAL** comes down to **precision**, **storage mechanics**, and **whether they can hold fractional values**.

**1. INT (Integer)**
* **What it is:** Exact whole numbers only. It cannot store decimal points or fractions.
* **Storage:** Highly efficient and takes up a fixed amount of space (usually 4 bytes).
* **Best Used For:** Counts, quantities, age, or surrogate/primary keys (e.g., `Employee_ID = 101`).

**2. FLOAT (Floating-Point)**
* **What it is:** An *approximate* numeric data type used to store numbers with decimals. It uses binary floating-point math (similar to scientific notation) to store extremely large or small numbers flexibly.
* **The Catch:** Because it is approximate, it can suffer from **rounding errors**. For example, storing `10.15` might actually save in the background as `10.149999999`.
* **Best Used For:** Scientific data, weather measurements, or physical dimensions where microscopic precision isn't critical, but a massive range of values is (e.g., the distance between planets, weight, speed).

**3. DECIMAL (also called NUMERIC)**
* **What it is:** An *exact* numeric data type for numbers with decimals. You define the exact total length (precision) and number of decimal places (scale) upfront. For example, `DECIMAL(10,2)` means a number with up to 10 total digits, 2 of which are after the decimal point.
* **The Catch:** It takes up more storage space and is slightly slower to calculate than `FLOAT`.
* **Best Used For:** Financial data, money, interest rates, or any calculation where rounding errors are completely unacceptable.

**Key Differences at a Glance**

| Feature | INT | FLOAT | DECIMAL / NUMERIC |
| --- | --- | --- | --- |
| **Data Type** | Exact Whole Number | Approximate Decimal | Exact Decimal |
| **Rounding Errors** | Never | **Yes** | Never |
| **Performance** | Fastest | Fast | Slower (requires more processing) |
| **Primary Use Case** | IDs, Counts | Scientific data, averages | Money, banking, exact metrics |

<br>

> **Interview Tip:** If an interviewer asks, *"Which data type would you use to store a currency value like a product price or an employee salary?"* **Never say FLOAT.** Always answer **DECIMAL** (or `NUMERIC`), explaining that `FLOAT` introduces floating-point rounding errors that will cause accounting discrepancies over time.

#### **16. What is NULL?**
In SQL, **NULL** is a special marker used to indicate that a data value does not exist in the database. It represents a state of being **unknown, missing, or inapplicable**.

For an interview, the most important thing to demonstrate is that you understand what `NULL` is *not*.

* `NULL` is **not zero** (`0`). Zero is a specific numerical value.
* `NULL` is **not an empty string** (`""` or `" "`). An empty string is a known text value with a length of zero.
* `NULL` means "we don't have this information."

**Core Rules of NULL**
* **Comparisons:** You cannot use standard mathematical operators (`=`, `<`, `>`, `!=`) to check for `NULL`. Because `NULL` is unknown, the database cannot evaluate if `NULL = NULL` (an unknown cannot equal an unknown). Instead, you must use the special operators **`IS NULL`** and **`IS NOT NULL`**.
* **Math Operations:** Any mathematical calculation involving a `NULL` results in `NULL` (e.g., `50 + NULL = NULL`).
* **Aggregate Functions:** Most aggregate functions (like `SUM()`, `AVG()`, `MAX()`) completely ignore `NULL` values when performing their calculations.

> **Interview Tip:** Interviewers often test your practical ability to handle `NULL`s. If they ask, *"How do you replace a NULL value with a default value (like '0' or 'Unknown') in a SELECT query?"*, you should mention the **`COALESCE()`** function (e.g., `COALESCE(Salary, 0)`), which returns the first non-null value in a list. You might also mention `IFNULL()` or `ISNULL()` depending on the specific SQL dialect.

#### **17. What is the difference between NULL and 0?**
The easiest way to explain the difference in an interview is to contrast a **known quantity** with an **unknown state**.

* **`0` (Zero)** is a specific, known, mathematical value. It means "exactly none" or a measured quantity of zero.
* **`NULL`** is a marker for missing, unknown, or inapplicable data. It means "we do not know what this value is."

**The Bank Account Analogy**
* If your account balance is **`0`**, the bank knows exactly how much money you have: none.
* If your account balance is **`NULL`**, the bank's system doesn't know your balance. Maybe the account creation is pending, or the system is experiencing an error.

**Key Differences at a Glance**

| Feature | `0` (Zero) | `NULL` |
| --- | --- | --- |
| **Meaning** | A specific numerical quantity. | Missing or unknown data. |
| **Comparisons** | Can use standard operators (`=`, `!=`, `<`, `>`). | Must use special operators (`IS NULL`, `IS NOT NULL`). |
| **Math Operations** | Calculates normally (e.g., `5 + 0 = 5`). | Results in an unknown (e.g., `5 + NULL = NULL`). |
| **Storage** | Takes up standard numeric storage space (e.g., 4 bytes for INT). | Stored using a metadata flag indicating absence. |

<br>

> **Interview Tip:** The most common technical trap regarding this difference involves the **`AVG()`** function.
> * If you have values `(10, 0, 20)`, the `AVG()` is **`10`** (30 divided by 3 rows).
> * If you have values `(10, NULL, 20)`, the `AVG()` is **`15`** (30 divided by 2 rows, because aggregate functions ignore `NULL` entirely). Knowing this distinction proves you have practical query experience.

#### **18. What is the purpose of the SELECT statement?**
The **`SELECT`** statement is the most frequently used command in SQL. Its primary purpose is to **retrieve (or fetch) data** from one or more tables in a database. When you execute a `SELECT` statement, the database engine returns the requested data in a temporary, grid-like format known as a **result set**.

**Core Capabilities**
* **Read-Only:** It is strictly used to view data. It never modifies, updates, or deletes the underlying data in the tables.
* **Column Selection:** You can specify exactly which columns you want to see (e.g., `SELECT First_Name, Salary FROM Employees`) or retrieve all columns at once using an asterisk (e.g., `SELECT * FROM Employees`).
* **Foundation for Filtering & Shaping:** The `SELECT` statement rarely works alone in a real-world scenario. It acts as the starting point and is combined with other clauses to shape the output:
* **`WHERE`**: To filter specific rows.
* **`JOIN`**: To combine data from multiple tables.
* **`GROUP BY`**: To aggregate data (like finding totals or averages).
* **`ORDER BY`**: To sort the final result set.

> **Interview Tip:** A very common interview question is: *"Why is it considered bad practice to use `SELECT *` in production code?"*
> You should answer that `SELECT *` retrieves every single column, which wastes network bandwidth, consumes unnecessary database memory, and can break application code if a new column is later added to the table. Always explicitly name the columns you need (`SELECT ID, Name`), even if you currently need all of them.

#### **19. What is the execution order of an SQL query?**
One of the most important concepts to master for SQL interviews is that the **logical execution order** (how the database engine actually processes the query) is completely different from the **lexical order** (how you type the query from top to bottom).

**The Logical Execution Order**  
When the database engine receives a standard query, it processes the clauses in this specific sequence:
1. **`FROM` (and `JOIN`)**: Identifies the working dataset by locating the base tables and merging them.
2. **`WHERE`**: Filters the raw rows based on specified conditions.
3. **`GROUP BY`**: Groups the remaining filtered rows based on common column values.
4. **`HAVING`**: Filters the grouped and aggregated data.
5. **`SELECT`**: Evaluates the expressions and determines exactly which columns to return to the screen.
6. **`DISTINCT`**: Removes any duplicate rows from the selected columns.
7. **`ORDER BY`**: Sorts the final result set.
8. **`LIMIT` / `OFFSET`**: Restricts the final number of rows returned to the user.

> **Interview Tip:** Interviewers constantly test this knowledge by asking: *"Why do I get an error if I use a column alias in the `WHERE` clause?"*
> Your answer: Because of the execution order. The `WHERE` clause runs (Step 2) *before* the `SELECT` clause (Step 5). Since the alias is created during the `SELECT` step, the database engine literally does not know the alias exists when it is trying to filter rows in the `WHERE` step.

#### **20. What is SQL injection?**
**SQL Injection (SQLi)** is a critical web security vulnerability that allows an attacker to interfere with the queries an application makes to its database.

It occurs when an application takes untrusted user input (like text entered into a login form or a search bar) and drops it directly into a database query without proper validation or sanitization.

**How It Works**  
Instead of entering a normal username, an attacker inputs malicious SQL code. Because the application doesn't separate the *user data* from the *SQL commands*, the database engine gets confused and executes the attacker's code.

**The Classic Example:**  
Imagine a login query:
`SELECT * FROM Users WHERE Username = '$user_input' AND Password = '$password_input'`

If an attacker enters `' OR '1'='1` into the username field, the query becomes:
`SELECT * FROM Users WHERE Username = '' OR '1'='1' AND Password = ''`

Because `1=1` is always true, the database returns a valid record (often the first user in the table, which is usually the admin), allowing the attacker to log in without a password.

**The Impact**
* **Data Breach:** Stealing sensitive information like passwords, credit card numbers, or personal data.
* **Data Manipulation:** Modifying or deleting records (e.g., dropping entire tables).
* **Authentication Bypass:** Logging in as an administrator.

**How to Prevent It (Crucial for Interviews)**
* **Parameterized Queries (Prepared Statements):** This is the **number one defense**. It forces the database to treat the user input strictly as a literal value (data), never as executable code.
* **Use of ORMs (Object-Relational Mapping):** Modern frameworks (like Hibernate, Entity Framework, or Django) generally use parameterized queries by default.
* **Input Validation & Sanitization:** Ensuring input matches expected formats (e.g., an email field only accepts a valid email format) and stripping out harmful SQL characters.

> **Interview Tip:** If asked, *"How do you prevent SQL injection?"*, do not just say "sanitize inputs." Your immediate and primary answer should always be **Parameterized Queries (or Prepared Statements)**. Interviewers want to hear this specific term because sanitization alone is prone to human error, whereas parameterized queries fix the root cause of the vulnerability.

#### **21. How do you filter rows using WHERE?**
The **`WHERE`** clause is used in SQL to filter records. It evaluates each row in a table and extracts only those rows that fulfill a specific condition, returning a smaller, targeted dataset.

We are applying examples from your `ecommerce.db` database:

**1. Single Condition Filtering**  
You use standard mathematical operators (`=`, `>`, `<`, `>=`, `<=`, `!=` or `<>`) to filter based on a single rule.

**2. Multiple Conditions (`AND`, `OR`)**  
You can chain multiple conditions together to make your filter more specific or more broad.

**3. Filtering with Special Operators**  
SQL provides specialized operators to make complex filtering easier to write:
* **`IN`**: Checks if a value matches any value in a list (cleaner than using multiple `OR` statements).
* **`BETWEEN`**: Filters values within a specific range (inclusive).
* **`LIKE`**: Performs pattern matching using wildcards (`%` for multiple characters, `_` for a single character).

> **Interview Tip:** Interviewers often ask about query optimization. If they ask, *"How can you make a slow query run faster?"*, one of the first things you should mention is **filtering early**. Applying a highly selective `WHERE` clause reduces the number of rows the database has to hold in memory before it performs heavy operations like `JOIN`s or `GROUP BY`s.

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("ecommerce.db")

# Find all products that cost more than $100
query = """
SELECT * 
FROM products 
WHERE Discount < 20;
"""
print(f"{query} \n {100*"#"}")
print(pd.read_sql(query, conn))
print(f"\n {100*"#"}")

# Find transactions that are completed AND are over $50
query = """
SELECT * 
FROM transactions 
WHERE City = 'Delhi' AND Tax < 500.0;
"""
print(f"{query} \n {100*"#"}")
print(pd.read_sql(query, conn))
print(f"\n {100*"#"}")

# Find employees who work in Sales OR Marketing
query = """
SELECT * 
FROM employees 
WHERE department = 'Sales' OR department = 'Marketing';
"""
print(f"{query} \n {100*"#"}")
print(pd.read_sql(query, conn))
print(f"\n {100*"#"}")

query = "SELECT * FROM customers WHERE city IN ('Kochi', 'Delhi', 'Jaipur');"
print(f"{query} \n {100*"#"}")
print(pd.read_sql(query, conn))
print(f"\n {100*"#"}")

query = "SELECT * FROM products WHERE StockQuantity BETWEEN 40 AND 50;"
print(f"{query} \n {100*"#"}")
print(pd.read_sql(query, conn))
print(f"\n {100*"#"}")

# Find students whose first name starts with 'A'
query = "SELECT * FROM students WHERE StudentName LIKE 'A%';"
print(f"{query} \n {100*"#"}")
print(pd.read_sql(query, conn))
print(f"\n {100*"#"}")
```

#### **22. What is the difference between WHERE and HAVING?**
The difference between **`WHERE`** and **`HAVING`** is one of the most frequently asked SQL interview questions. While both are used to filter data, the critical difference is **when** they apply that filter during the query's execution.
* **`WHERE`** filters individual rows **before** any grouping or aggregation takes place.
* **`HAVING`** filters the aggregated groups **after** the `GROUP BY` clause has been applied.

**Key Differences at a Glance**

| Feature | `WHERE` Clause | `HAVING` Clause |
| --- | --- | --- |
| **Execution Order** | Runs early (before `GROUP BY`). | Runs late (after `GROUP BY`). |
| **Target** | Filters individual, raw rows. | Filters grouped rows. |
| **Aggregate Functions** | **Cannot** contain aggregate functions (`SUM`, `COUNT`, `AVG`). | **Can** (and usually does) contain aggregate functions. |

**Example in Action (Using `ecommerce.db`)**  
Imagine you want to find out which departments in your `employees` table have a total payroll of more than $100,000, but you only want to count active employees. You must use **both** clauses to achieve this.

If you tried to write `WHERE SUM(salary) > 100000`, the database would throw an error because the `WHERE` clause runs before the `SUM()` has even been calculated.

> **Interview Tip:** A great way to impress an interviewer is to discuss **performance**. Mention that it is best practice to push as much filtering as possible into the `WHERE` clause rather than the `HAVING` clause. Filtering out rows early (with `WHERE`) means the database has to hold fewer rows in memory when performing the computationally expensive `GROUP BY` operation.

```python
# 1. Filters raw rows: Keeps only active employees
# 2. Groups the remaining rows by department
# 3. Filters groups: Keeps only departments with high payroll
query = """
SELECT Department, SUM(Salary) AS Total_Payroll
FROM employees
WHERE Gender = 'Male'
GROUP BY Department
HAVING SUM(Salary) > 100000;
"""
print(f"{query} \n {100*"#"}")
print(pd.read_sql(query, conn))
print(f"\n {100*"#"}")
```
