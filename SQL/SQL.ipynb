## Connecting SQL in Jupyter using SQLite3

```python
import sqlite3
import pandas as pd

# Create a database
conn = sqlite3.connect("ecommerce.db")
```

```python
# Load CSV files
path = "./../Datasets/"
customers = pd.read_csv(path + "customers.csv")
products = pd.read_csv(path + "products.csv")
employees = pd.read_csv(path + "employees.csv")
transactions = pd.read_csv(path + "transactions.csv")
students = pd.read_csv(path + "students.csv")

# Write DataFrames into SQL tables
customers.to_sql("customers", conn, if_exists="replace", index=False)
products.to_sql("products", conn, if_exists="replace", index=False)
employees.to_sql("employees", conn, if_exists="replace", index=False)
transactions.to_sql("transactions", conn, if_exists="replace", index=False)
students.to_sql("students", conn, if_exists="replace", index=False)
```

```python
# Execute SQL queries (Example)
query = """
SELECT City, COUNT(*) AS total_customers
FROM customers
GROUP BY City
ORDER BY total_customers DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT * FROM customers;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT * FROM products;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT * FROM employees;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT * FROM transactions;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT * FROM students;
"""

pd.read_sql(query, conn)
```
