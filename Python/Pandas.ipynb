```python
import pandas as pd
import numpy as np

# Reproducibility
np.random.seed(42)

n = 5000

# Sample values
customer_ids = np.random.randint(1000, 1300, n)
order_ids = np.arange(10001, 10001 + n)

names = ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Helen", "Ivy", "Jack"]
cities = ["New York", "Chicago", "Dallas", "Boston","Seattle", "San Francisco", "Miami", "Houston"]
products = ["Laptop", "Mouse", "Keyboard", "Monitor", "Phone", "Tablet", "Headphones", "Printer"]
categories = ["Electronics", "Accessories", "Office", "Mobile"]
payment_methods = ["Credit Card", "Debit Card", "UPI", "Cash", "Net Banking"]
genders = ["Male", "Female"]

# Dates
dates = pd.date_range("2023-01-01", periods=730)
order_dates = np.random.choice(dates, n)
quantity = np.random.randint(1, 8, n)
price = np.random.randint(100, 5000, n)
discount = np.random.choice([0, 5, 10, 15, 20], n)
sales = quantity * price * (1 - discount / 100)
ratings = np.random.randint(1, 6, n)
age = np.random.randint(18, 70, n)

# DataFrame
df = pd.DataFrame({
    "OrderID": order_ids,
    "CustomerID": customer_ids,
    "CustomerName": np.random.choice(names, n),
    "Age": age,
    "Gender": np.random.choice(genders, n),
    "City": np.random.choice(cities, n),
    "Product": np.random.choice(products, n),
    "Category": np.random.choice(categories, n),
    "Quantity": quantity,
    "Price": price,
    "Discount": discount,
    "Sales": sales.round(2),
    "Rating": ratings,
    "PaymentMethod": np.random.choice(payment_methods, n),
    "OrderDate": order_dates
})

# -----------------------------
# Introduce Missing Values
# -----------------------------
for col in ["Age", "City", "Rating"]:
    idx = np.random.choice(df.index, 120, replace=False)
    df.loc[idx, col] = np.nan

# -----------------------------
# Introduce Duplicate Rows
# -----------------------------
duplicates = df.sample(50, random_state=1)
df = pd.concat([df, duplicates], ignore_index=True)

# Shuffle dataset
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Save CSV
df.to_csv("sales_data.csv", index=False)
print(df.head())
print("\nDataset Shape:", df.shape)
print("\nCSV saved as sales_data.csv")
```

```python
import pandas as pd
import numpy as np
import random

# ---------------------------------------------------
# Configuration
# ---------------------------------------------------
np.random.seed(42)
random.seed(42)
NUM_PRODUCTS = 2000

# ---------------------------------------------------
# Categories
# ---------------------------------------------------
categories = {
    "Electronics": ["Laptop", "Desktop", "Monitor", "Mouse", "Keyboard", "Printer", "Router", "Speaker"],
    "Mobile": ["Smartphone", "Tablet", "Smartwatch", "Power Bank", "Charger", "Earbuds"],
    "Home Appliances": ["Microwave", "Refrigerator", "Air Conditioner", "Vacuum Cleaner", "Ceiling Fan", "Mixer Grinder"],
    "Furniture": ["Chair", "Table", "Sofa", "Wardrobe", "Bookshelf", "Office Desk"],
    "Fashion": ["T-Shirt", "Jeans", "Jacket", "Sneakers", "Watch", "Backpack"],
    "Sports": ["Football", "Cricket Bat", "Badminton Racket", "Yoga Mat", "Dumbbell"],
    "Books": ["Novel", "Python Book", "SQL Book", "AI Book", "Data Science Book"]
}

# ---------------------------------------------------
# Brands
# ---------------------------------------------------
brands = [
    "Apple", "Samsung", "Dell", "HP", "Lenovo", "Sony", "LG", "Nike", "Adidas", "Puma",
    "Boat", "Asus", "Canon", "Logitech", "Philips", "Bajaj", "Havells", "Mi", "OnePlus", "Realme"
]

# ---------------------------------------------------
# Supplier IDs
# ---------------------------------------------------
supplier_ids = np.arange(5001, 5301)

# ---------------------------------------------------
# Product Generation
# ---------------------------------------------------
records = []
launch_dates = pd.date_range(start="2019-01-01", end="2025-12-31")

for pid in range(100001, 100001 + NUM_PRODUCTS):
    category = random.choice(list(categories.keys()))
    product = random.choice(categories[category])
    brand = random.choice(brands)
    product_name = f"{brand} {product}"
    cost_price = random.randint(100, 50000)
    margin = random.uniform(0.10, 0.60)
    selling_price = round(cost_price * (1 + margin), 2)
    stock = random.randint(0, 500)
    supplier = random.choice(supplier_ids)
    rating = round(random.uniform(2.5, 5.0), 1)
    warranty = random.choice(["No Warranty", "6 Months", "1 Year", "2 Years", "3 Years"])
    weight = round(random.uniform(0.2, 30), 2)
    launch = random.choice(launch_dates)

    records.append([pid, product_name, category, brand, supplier, cost_price, selling_price, stock, rating, warranty, weight, launch])

# ---------------------------------------------------
# DataFrame
# ---------------------------------------------------
products = pd.DataFrame(records, 
                        columns = [
                                "ProductID", "ProductName", "Category", "Brand", "SupplierID", "CostPrice",
                                "SellingPrice", "StockQuantity", "Rating", "Warranty", "Weight(kg)", "LaunchDate"
                            ])

# ---------------------------------------------------
# Profit Column
# ---------------------------------------------------
products["Profit"] = (products["SellingPrice"]-products["CostPrice"]).round(2)

# ---------------------------------------------------
# Discount Column
# ---------------------------------------------------
products["Discount"] = np.random.choice([0, 5, 10, 15, 20, 25, 30], len(products))

# ---------------------------------------------------
# Missing Values
# ---------------------------------------------------
for col in ["Rating", "Warranty", "Weight(kg)"]:
    idx = np.random.choice(products.index, 40, replace=False)
    products.loc[idx, col] = np.nan

# ---------------------------------------------------
# Duplicate Rows
# ---------------------------------------------------
duplicates = products.sample(25, random_state=42)
products = pd.concat([products, duplicates], ignore_index=True)

# ---------------------------------------------------
# Shuffle Dataset
# ---------------------------------------------------
products = products.sample(frac=1, random_state=42).reset_index(drop=True)

# ---------------------------------------------------
# Save CSV
# ---------------------------------------------------
products.to_csv("products.csv", index=False)

# ---------------------------------------------------
# Output
# ---------------------------------------------------
print(products.head())
print("\nShape:", products.shape)
print("\nproducts.csv created successfully!")
```

```python
import pandas as pd
import numpy as np
import random

# -------------------------------------------------
# Configuration
# -------------------------------------------------
np.random.seed(42)
random.seed(42)
NUM_EMPLOYEES = 500

# -------------------------------------------------
# Sample Data
# -------------------------------------------------
first_names = [
    "Aarav","Vivaan","Aditya","Arjun","Sai","Ananya","Diya","Isha","Priya","Riya","Rahul","Rohan","Karan",
    "Neha","Pooja","Sneha","Amit","Akash","Meera","Sonia","Rakesh","Deepak","Nikhil","Komal","Varun"
]

last_names = [
    "Sharma","Patel","Singh","Gupta","Das","Roy","Banerjee","Verma",
    "Nair","Joshi","Mehta","Kulkarni","Reddy","Yadav","Khan"
]

departments = ["Sales", "Customer Support", "HR", "Finance", "IT", "Marketing", "Operations", "Warehouse"]

designations = {
    "Sales": ["Sales Executive","Senior Sales Executive","Sales Manager"],
    "Customer Support": ["Support Executive","Senior Support Executive","Support Manager"],
    "HR": ["HR Executive","HR Manager"],
    "Finance": ["Accountant","Finance Analyst","Finance Manager"],
    "IT": ["Software Engineer","Senior Software Engineer","Tech Lead"],
    "Marketing": ["Marketing Executive","Digital Marketer","Marketing Manager"],
    "Operations": ["Operations Executive","Operations Manager"],
    "Warehouse": ["Warehouse Associate","Warehouse Supervisor"]
}

cities = {
    "Delhi":"Delhi",
    "Mumbai":"Maharashtra",
    "Bengaluru":"Karnataka",
    "Chennai":"Tamil Nadu",
    "Hyderabad":"Telangana",
    "Kolkata":"West Bengal",
    "Pune":"Maharashtra",
    "Ahmedabad":"Gujarat",
    "Jaipur":"Rajasthan",
    "Lucknow":"Uttar Pradesh"
}

genders = ["Male","Female"]

# -------------------------------------------------
# Generate Data
# -------------------------------------------------
records = []
employee_ids = np.arange(2001, 2001 + NUM_EMPLOYEES)
hire_dates = pd.date_range(start="2015-01-01", end="2025-12-31")

for emp_id in employee_ids:
    first = random.choice(first_names)
    last = random.choice(last_names)
    name = f"{first} {last}"
    gender = random.choice(genders)
    age = random.randint(21, 60)
    department = random.choice(departments)
    designation = random.choice(designations[department])
    experience = round(random.uniform(0.5, 25),1)
    salary = random.randint(300000,1800000)
    bonus = round(salary * random.uniform(0.05,0.25),2)
    city = random.choice(list(cities.keys()))
    state = cities[city]
    performance = round(random.uniform(2.5,5.0),1)
    hire_date = random.choice(hire_dates)

    # About 15% of employees have no manager (top-level roles)
    if random.random() < 0.15:
        manager = np.nan
    else:
        manager = random.choice(employee_ids)
        # Avoid self-manager
        while manager == emp_id:
            manager = random.choice(employee_ids)

    email = (first.lower() + "." + last.lower() + str(emp_id) + "@company.com")
    phone = "9" + "".join(random.choices("0123456789",k=9))

    records.append([
        emp_id, name, gender, age, department, designation, experience, salary, 
        bonus, city, state, "India", manager, performance, hire_date, email, phone
    ])

# -------------------------------------------------
# DataFrame
# -------------------------------------------------
employees = pd.DataFrame(records, 
                columns=[
                "EmployeeID", "EmployeeName", "Gender", "Age", "Department", "Designation", "Experience", "Salary", 
                "Bonus", "City", "State", "Country", "ManagerID", "PerformanceRating", "HireDate", "Email", "Phone"
])

# -------------------------------------------------
# Annual Compensation
# -------------------------------------------------
employees["AnnualCompensation"] = (employees["Salary"] + employees["Bonus"]).round(2)

# -------------------------------------------------
# Missing Values
# -------------------------------------------------
for col in ["Age", "PerformanceRating", "Phone"]:
    idx = np.random.choice(employees.index, 20, replace=False)
    employees.loc[idx,col] = np.nan

# -------------------------------------------------
# Duplicate Rows
# -------------------------------------------------
duplicates = employees.sample(10, random_state=42)
employees = pd.concat([employees, duplicates], ignore_index=True)

# -------------------------------------------------
# Shuffle
# -------------------------------------------------
employees = employees.sample(frac=1, random_state=42).reset_index(drop=True)

# -------------------------------------------------
# Save CSV
# -------------------------------------------------
employees.to_csv("employees.csv", index=False)
print(employees.head())
print("\nShape:", employees.shape)
print("\nemployees.csv created successfully!")
```

```python
import pandas as pd
import numpy as np
import random

# -----------------------------------------
# Configuration
# -----------------------------------------
np.random.seed(42)
random.seed(42)

NUM_STUDENTS = 5000

# -----------------------------------------
# Sample Data
# -----------------------------------------

first_names = [
    "Aarav","Vivaan","Aditya","Arjun","Sai","Ayaan",
    "Ananya","Diya","Isha","Priya","Riya","Sneha",
    "Rahul","Rohan","Karan","Amit","Akash","Meera",
    "Sonia","Pooja","Nikhil","Varun","Komal","Neha"
]

last_names = [
    "Sharma","Patel","Singh","Gupta","Das",
    "Roy","Banerjee","Verma","Nair","Joshi",
    "Reddy","Khan","Mehta","Kulkarni","Yadav"
]

cities = {
    "Delhi":"Delhi",
    "Mumbai":"Maharashtra",
    "Bengaluru":"Karnataka",
    "Chennai":"Tamil Nadu",
    "Hyderabad":"Telangana",
    "Kolkata":"West Bengal",
    "Pune":"Maharashtra",
    "Ahmedabad":"Gujarat",
    "Lucknow":"Uttar Pradesh",
    "Jaipur":"Rajasthan"
}

courses = [
    "Computer Science",
    "Information Technology",
    "Electronics",
    "Mechanical",
    "Civil",
    "Electrical",
    "Mathematics",
    "Physics",
    "Chemistry",
    "Business Administration"
]

years = [1, 2, 3, 4]

semesters = [1,2,3,4,5,6,7,8]

genders = ["Male","Female"]

# -----------------------------------------
# Generate Student Records
# -----------------------------------------

records = []

student_ids = np.arange(100001, 100001 + NUM_STUDENTS)

admission_dates = pd.date_range(
    start="2020-01-01",
    end="2025-12-31"
)

for sid in student_ids:

    first = random.choice(first_names)
    last = random.choice(last_names)

    name = f"{first} {last}"

    gender = random.choice(genders)

    age = random.randint(17,28)

    city = random.choice(list(cities.keys()))
    state = cities[city]

    course = random.choice(courses)

    year = random.choice(years)

    semester = random.choice(semesters)

    attendance = round(random.uniform(60,100),2)

    maths = random.randint(35,100)
    physics = random.randint(35,100)
    programming = random.randint(35,100)

    average = round(
        (maths + physics + programming)/3,
        2
    )

    if average >= 90:
        grade = "A+"
    elif average >= 80:
        grade = "A"
    elif average >= 70:
        grade = "B"
    elif average >= 60:
        grade = "C"
    elif average >= 50:
        grade = "D"
    else:
        grade = "F"

    admission = random.choice(admission_dates)

    email = (
        first.lower() +
        "." +
        last.lower() +
        str(sid) +
        "@student.edu"
    )

    phone = "9" + "".join(
        random.choices("0123456789", k=9)
    )

    scholarship = random.choice(
        ["Yes","No"]
    )

    records.append([
        sid,
        name,
        gender,
        age,
        city,
        state,
        "India",
        course,
        year,
        semester,
        attendance,
        maths,
        physics,
        programming,
        average,
        grade,
        scholarship,
        admission,
        email,
        phone
    ])

# -----------------------------------------
# Create DataFrame
# -----------------------------------------

students = pd.DataFrame(records, columns=[
    "StudentID",
    "StudentName",
    "Gender",
    "Age",
    "City",
    "State",
    "Country",
    "Course",
    "Year",
    "Semester",
    "Attendance",
    "MathMarks",
    "PhysicsMarks",
    "ProgrammingMarks",
    "AverageMarks",
    "Grade",
    "Scholarship",
    "AdmissionDate",
    "Email",
    "Phone"
])

# -----------------------------------------
# Introduce Missing Values
# -----------------------------------------

for col in [
    "Age",
    "Attendance",
    "Phone"
]:
    idx = np.random.choice(
        students.index,
        80,
        replace=False
    )

    students.loc[idx,col] = np.nan

# -----------------------------------------
# Duplicate Rows
# -----------------------------------------

duplicates = students.sample(
    50,
    random_state=42
)

students = pd.concat(
    [students,duplicates],
    ignore_index=True
)

# -----------------------------------------
# Shuffle
# -----------------------------------------

students = students.sample(
    frac=1,
    random_state=42
).reset_index(drop=True)

# -----------------------------------------
# Save CSV
# -----------------------------------------

students.to_csv(
    "students.csv",
    index=False
)

print(students.head())

print("\nDataset Shape:", students.shape)

print("\nstudents.csv created successfully!")
```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# ----------------------------------------------------
# Configuration
# ----------------------------------------------------
np.random.seed(42)
random.seed(42)

NUM_TRANSACTIONS = 100000

# ----------------------------------------------------
# Sample Data
# ----------------------------------------------------

payment_methods = [
    "Credit Card",
    "Debit Card",
    "UPI",
    "Net Banking",
    "Cash",
    "Wallet"
]

transaction_types = [
    "Purchase",
    "Refund",
    "Subscription",
    "Transfer",
    "Withdrawal",
    "Deposit"
]

status_list = [
    "Success",
    "Pending",
    "Failed"
]

currencies = [
    "INR",
    "USD",
    "EUR"
]

merchant_categories = [
    "Electronics",
    "Fashion",
    "Groceries",
    "Travel",
    "Healthcare",
    "Restaurants",
    "Entertainment",
    "Education",
    "Fuel",
    "Utilities"
]

cities = [
    "Delhi",
    "Mumbai",
    "Bengaluru",
    "Hyderabad",
    "Kolkata",
    "Chennai",
    "Pune",
    "Ahmedabad",
    "Lucknow",
    "Jaipur"
]

devices = [
    "Android",
    "iPhone",
    "Windows",
    "Mac",
    "POS Terminal"
]

# ----------------------------------------------------
# Generate Data
# ----------------------------------------------------

records = []

start_date = datetime(2022, 1, 1)
end_date = datetime(2025, 12, 31)

total_seconds = int((end_date - start_date).total_seconds())

for tid in range(1, NUM_TRANSACTIONS + 1):

    customer_id = random.randint(100001, 110000)

    amount = round(np.random.gamma(3, 1500), 2)

    payment_method = random.choice(payment_methods)

    transaction_type = random.choice(transaction_types)

    status = random.choices(
        status_list,
        weights=[92, 5, 3],
        k=1
    )[0]

    currency = random.choices(
        currencies,
        weights=[90, 7, 3],
        k=1
    )[0]

    merchant = random.choice(merchant_categories)

    city = random.choice(cities)

    device = random.choice(devices)

    timestamp = start_date + timedelta(
        seconds=random.randint(0, total_seconds)
    )

    # Fraud (about 2%)
    fraud = random.choices(
        [0, 1],
        weights=[98, 2],
        k=1
    )[0]

    # Cashback
    cashback = round(amount * random.uniform(0, 0.10), 2)

    # Tax (GST-like)
    tax = round(amount * 0.18, 2)

    # Processing fee
    processing_fee = round(amount * random.uniform(0.005, 0.02), 2)

    records.append([
        tid,
        customer_id,
        timestamp,
        transaction_type,
        payment_method,
        merchant,
        amount,
        tax,
        processing_fee,
        cashback,
        currency,
        status,
        city,
        device,
        fraud
    ])

# ----------------------------------------------------
# Create DataFrame
# ----------------------------------------------------

transactions = pd.DataFrame(records, columns=[
    "TransactionID",
    "CustomerID",
    "TransactionTime",
    "TransactionType",
    "PaymentMethod",
    "MerchantCategory",
    "Amount",
    "Tax",
    "ProcessingFee",
    "Cashback",
    "Currency",
    "Status",
    "City",
    "Device",
    "FraudFlag"
])

# ----------------------------------------------------
# Net Amount
# ----------------------------------------------------

transactions["NetAmount"] = (
    transactions["Amount"]
    + transactions["Tax"]
    + transactions["ProcessingFee"]
    - transactions["Cashback"]
).round(2)

# ----------------------------------------------------
# Missing Values
# ----------------------------------------------------

for col in ["City", "Device", "PaymentMethod"]:
    idx = np.random.choice(
        transactions.index,
        500,
        replace=False
    )
    transactions.loc[idx, col] = np.nan

# ----------------------------------------------------
# Duplicate Records
# ----------------------------------------------------

duplicates = transactions.sample(
    200,
    random_state=42
)

transactions = pd.concat(
    [transactions, duplicates],
    ignore_index=True
)

# ----------------------------------------------------
# Shuffle Dataset
# ----------------------------------------------------

transactions = transactions.sample(
    frac=1,
    random_state=42
).reset_index(drop=True)

# ----------------------------------------------------
# Save CSV
# ----------------------------------------------------

transactions.to_csv(
    "transactions.csv",
    index=False
)

print(transactions.head())

print("\nShape:", transactions.shape)

print("\ntransactions.csv created successfully!")
```
