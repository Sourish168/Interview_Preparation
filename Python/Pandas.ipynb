## Dataset Creation

```python
import pandas as pd
import numpy as np

# Reproducibility
np.random.seed(42)

n = 5000

# Sample values
customer_ids = np.random.randint(1000, 1300, n)
order_ids = np.arange(10001, 10001 + n)

names = ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Helen", "Ivy", "Jack"]
cities = ["New York", "Chicago", "Dallas", "Boston","Seattle", "San Francisco", "Miami", "Houston"]
products = ["Laptop", "Mouse", "Keyboard", "Monitor", "Phone", "Tablet", "Headphones", "Printer"]
categories = ["Electronics", "Accessories", "Office", "Mobile"]
payment_methods = ["Credit Card", "Debit Card", "UPI", "Cash", "Net Banking"]
genders = ["Male", "Female"]

# Dates
dates = pd.date_range("2023-01-01", periods=730)
order_dates = np.random.choice(dates, n)
quantity = np.random.randint(1, 8, n)
price = np.random.randint(100, 5000, n)
discount = np.random.choice([0, 5, 10, 15, 20], n)
sales = quantity * price * (1 - discount / 100)
ratings = np.random.randint(1, 6, n)
age = np.random.randint(18, 70, n)

# DataFrame
df = pd.DataFrame({
    "OrderID": order_ids,
    "CustomerID": customer_ids,
    "CustomerName": np.random.choice(names, n),
    "Age": age,
    "Gender": np.random.choice(genders, n),
    "City": np.random.choice(cities, n),
    "Product": np.random.choice(products, n),
    "Category": np.random.choice(categories, n),
    "Quantity": quantity,
    "Price": price,
    "Discount": discount,
    "Sales": sales.round(2),
    "Rating": ratings,
    "PaymentMethod": np.random.choice(payment_methods, n),
    "OrderDate": order_dates
})

# -----------------------------
# Introduce Missing Values
# -----------------------------
for col in ["Age", "City", "Rating"]:
    idx = np.random.choice(df.index, 120, replace=False)
    df.loc[idx, col] = np.nan

# -----------------------------
# Introduce Duplicate Rows
# -----------------------------
duplicates = df.sample(50, random_state=1)
df = pd.concat([df, duplicates], ignore_index=True)

# Shuffle dataset
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Save CSV
df.to_csv("sales_data.csv", index=False)
print(df.head())
print("\nDataset Shape:", df.shape)
print("\nCSV saved as sales_data.csv")
```
