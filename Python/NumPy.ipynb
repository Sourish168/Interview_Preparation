#### 1. What is NumPy and why is it used?
- NumPy or "Numerical Python" is a python library used for numerical calculations in python. And NumPy is very fast in terms of numerical calculations. We can use NumPy in python as mentioned below.

```python
import numpy as np
```

#### 2. What is an ndarray?
- Array in NumPy is an one dimensional array. ndarray is n dimentional array in NumPy.

#### 3. Difference between Python list and NumPy array?
- The difference between Python list and NumPy array is mentioned below in terms of speed(NumPy Array is Faster) and element datatypes(NumPy array can only take numbers)

```python
%time np.array([1, 2, 3, 4])
%time list([1, 2, 3, 4])
```
