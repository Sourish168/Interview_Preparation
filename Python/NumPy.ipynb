#### 1. What is NumPy and why is it used?
- NumPy or "Numerical Python" is a python library used for numerical calculations in python. And NumPy is very fast in terms of numerical calculations. We can use NumPy in python as mentioned below.

```python
import numpy as np
```

#### 2. What is an ndarray?
- Array in NumPy is an one dimensional array. ndarray is n dimentional array in NumPy.

```python
a = np.ndarray(shape=(3, 4))
print(a)
```

#### 3. Difference between Python list and NumPy array?
- The main difference between Python list and NumPy array is mentioned below in terms of speed(NumPy Array is Faster) and element datatypes(NumPy array can only take numbers)

```python
%time np.array([1, 2, 3, 4])
%time list([1, 2, 3, 4])

%time np.array([1.0, 2.0, 3.0, 4.0])
%time list([1.0, 2.0, 3.0, 4.0])

%time np.array([1.0/2, 2.0/2, 3.0/2, 4.0/2])
%time list([1.0/2, 2.0/2, 3.0/2, 4.0/2])
```

```python
%time print(np.array([1, 2.5, "Data Science", {1: "toy", 2: "ball"}]))
%time print(list([1, 2.5, "Data Science", {1: "toy", 2: "ball"}]))
```

#### 4. How do you create a NumPy array?

```python
import numpy as np
a = np.array([1, 2, 5, 4.3, 33.0])
print(a, type(a))
```

#### 5. What is the difference between np.array() and np.asarray()?

```python
import numpy as np

original_array = np.array([1, 3, 5, 7])
new_array = np.array(original_array) # Always makes a copy in the memory
print(original_array)
print(new_array)

original_array[0] = 99
print(original_array)
print(new_array)
```

```python
import numpy as np

original_array = np.array([1, 3, 5, 7])
new_array = np.asarray(original_array) # Uses the same memory location as the original_array(the datatype should be same)
print(original_array)
print(new_array)

original_array[0] = 99
print(original_array)
print(new_array)
```

#### 6. How do you check the shape of an array?

```python
a = np.array([[1, 4, 7], [2, 5, 8], [3, 6, 9]])
print(a)
print(a.shape)
```
